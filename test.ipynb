```python
loanAmount=100000
loanInterestRate=5
loanTerm=5
```

本金攤還

```python
period=loanTerm*12
monthlyRate=0.1/12
principalRepay=int(loanAmount/period)
for i in range(period):
    Interest=round(loanAmount*monthlyRate)
    loanAmount-=principalRepay
    totalRepay=principalRepay+Interest
    if i==period-1:
        principalRepay+=loanAmount
        totalRepay=principalRepay+Interest
        print("<期別{:2}> 還款本金：{:5}/利息：{:4}/本利和：{:5}/貸款餘額：{:5}".format(i+1,principalRepay,Interest,totalRepay,0))
    else:
        print("<期別{:2}> 還款本金：{:5}/利息：{:4}/本利和：{:5}/貸款餘額：{:5}".format(i+1,principalRepay,Interest,totalRepay,loanAmount))
    print("---------------------------------------------------------------")

```
